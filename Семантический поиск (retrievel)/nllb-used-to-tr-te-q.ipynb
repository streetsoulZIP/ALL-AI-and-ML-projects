# Перевод на 9 языков

```python
import os, torch

import pandas as pd

os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "max_split_size_mb:128,expandable_segments:True,garbage_collection_threshold:0.6"
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cuda.enable_flash_sdp(True)
torch.backends.cuda.enable_mem_efficient_sdp(True)
torch.backends.cuda.enable_math_sdp(False)
```

```python
import bitsandbytes
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from typing import List, Dict


MODEL_NAME = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)


model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_NAME,
    device_map="auto",
    # load_in_8bit=True,          # int8 через bitsandbytes
    torch_dtype=torch.float16,  # вычисления вполнакала
    low_cpu_mem_usage=True
)


model.eval()


TARGET_LANGS = {
        "fra_Latn": "French 🇫🇷",
        "deu_Latn": "German 🇩🇪",
        "spa_Latn": "Spanish 🇪🇸",
        "ita_Latn": "Italian 🇮🇹",
        "pol_Latn": "Polish 🇵🇱",
        "swe_Latn": "Swedish 🇸🇪",
        "por_Latn": "Portuguese 🇵🇹",
        "ron_Latn": "Romanian 🇷🇴",
        "ukr_Cyrl": "Ukrainian 🇺🇦",
    }

def translater(model, tokenizer, df, TEXT_COL='queries', prefix_for_saver='queries_', TARGET_LANGS=TARGET_LANGS):


    # --- один раз, перед началом перевода ---
    device = model.get_encoder().embed_tokens.weight.device


    # >>> УКАЖИТЕ ЯЗЫК ИСТОЧНИКА <<<
    SRC_LANG = "eng_Latn"  # поменяйте при необходимости

    # 10 целевых языков по вашему списку

    tokenizer.src_lang = SRC_LANG


    texts = df[TEXT_COL].dropna().to_list()
    batch_size = 2048
    save_par_batch = 32

    for tgt_lang in TARGET_LANGS.keys():
        forced_bos_id = tokenizer.convert_tokens_to_ids(tgt_lang)
        tokenizer.tgt_lang = tgt_lang
        decoded_all = []
        num_batch = 0
        for i in range(0, len(texts), batch_size):
            num_batch+=1
            batch = texts[i:i+batch_size]

            enc = tokenizer(
                  batch,
                  return_tensors="pt",
                  padding=True,
                  truncation=True
              )
            # Перенос на девайс модели (device_map="auto" сам распределит веса;
            # но входы нужно кинуть на тот же первый девайс lm_head)
            enc = {k: v.to(device) for k, v in enc.items()}
            with torch.inference_mode():
                out = model.generate(
                    **enc,
                    forced_bos_token_id=forced_bos_id,
                    do_sample=False,
                    max_new_tokens=96
                )

            decoded = tokenizer.batch_decode(out, skip_special_tokens=True)
            decoded_all.extend(decoded)


            if num_batch%save_par_batch:
              out_df = pd.DataFrame({
                  f"{SRC_LANG}": texts[:len(decoded_all)],
                  f"{tgt_lang}": decoded_all,
              })
              out_path = f"{prefix_for_saver}{tgt_lang}.parquet"
              out_df.to_parquet(out_path, index=False)
        out_df.to_parquet(out_path, index=False)

```

```python
TARGET_LANGS = {
        # "fra_Latn": "French 🇫🇷",
        # "deu_Latn": "German 🇩🇪",
        # "spa_Latn": "Spanish 🇪🇸",
        "ita_Latn": "Italian 🇮🇹",
        "pol_Latn": "Polish 🇵🇱",
        "swe_Latn": "Swedish 🇸🇪",
        "por_Latn": "Portuguese 🇵🇹",
        "ron_Latn": "Romanian 🇷🇴",
        "ukr_Cyrl": "Ukrainian 🇺🇦",
    }
```

```python
q_for_tf = pd.read_parquet("/kaggle/input/queries/data_for_tf.parquet")
translater(model, tokenizer, q_for_tf, TEXT_COL='queries', prefix_for_saver='queries_tf_', TARGET_LANGS=TARGET_LANGS)
```
