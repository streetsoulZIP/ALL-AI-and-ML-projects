```python
import bitsandbytes
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from typing import List, Dict


MODEL_NAME = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)


model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_NAME,
    device_map="auto",
    # load_in_8bit=True,          # int8 Ñ‡ĞµÑ€ĞµĞ· bitsandbytes
    torch_dtype=torch.float16,  # Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ñ Ğ²Ğ¿Ğ¾Ğ»Ğ½Ğ°ĞºĞ°Ğ»Ğ°
    low_cpu_mem_usage=True
)


model.eval()


TARGET_LANGS = {
        "fra_Latn": "French ğŸ‡«ğŸ‡·",
        "deu_Latn": "German ğŸ‡©ğŸ‡ª",
        "spa_Latn": "Spanish ğŸ‡ªğŸ‡¸",
        "ita_Latn": "Italian ğŸ‡®ğŸ‡¹",
        "pol_Latn": "Polish ğŸ‡µğŸ‡±",
        "swe_Latn": "Swedish ğŸ‡¸ğŸ‡ª",
        "por_Latn": "Portuguese ğŸ‡µğŸ‡¹",
        "ron_Latn": "Romanian ğŸ‡·ğŸ‡´",
        "ukr_Cyrl": "Ukrainian ğŸ‡ºğŸ‡¦",
    }

def translater(model, tokenizer, df, TEXT_COL='queries', prefix_for_saver='queries_', TARGET_LANGS=TARGET_LANGS):


    # --- Ğ¾Ğ´Ğ¸Ğ½ Ñ€Ğ°Ğ·, Ğ¿ĞµÑ€ĞµĞ´ Ğ½Ğ°Ñ‡Ğ°Ğ»Ğ¾Ğ¼ Ğ¿ĞµÑ€ĞµĞ²Ğ¾Ğ´Ğ° ---
    device = model.get_encoder().embed_tokens.weight.device


    # >>> Ğ£ĞšĞĞ–Ğ˜Ğ¢Ğ• Ğ¯Ğ—Ğ«Ğš Ğ˜Ğ¡Ğ¢ĞĞ§ĞĞ˜ĞšĞ <<<
    SRC_LANG = "eng_Latn"  # Ğ¿Ğ¾Ğ¼ĞµĞ½ÑĞ¹Ñ‚Ğµ Ğ¿Ñ€Ğ¸ Ğ½ĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ÑÑ‚Ğ¸

    # 10 Ñ†ĞµĞ»ĞµĞ²Ñ‹Ñ… ÑĞ·Ñ‹ĞºĞ¾Ğ² Ğ¿Ğ¾ Ğ²Ğ°ÑˆĞµĞ¼Ñƒ ÑĞ¿Ğ¸ÑĞºÑƒ

    tokenizer.src_lang = SRC_LANG


    texts = df[TEXT_COL].dropna().to_list()
    batch_size = 2048
    save_par_batch = 32

    for tgt_lang in TARGET_LANGS.keys():
        forced_bos_id = tokenizer.convert_tokens_to_ids(tgt_lang)
        tokenizer.tgt_lang = tgt_lang
        decoded_all = []
        num_batch = 0
        for i in range(0, len(texts), batch_size):
            num_batch+=1
            batch = texts[i:i+batch_size]

            enc = tokenizer(
                  batch,
                  return_tensors="pt",
                  padding=True,
                  truncation=True
              )
            # ĞŸĞµÑ€ĞµĞ½Ğ¾Ñ Ğ½Ğ° Ğ´ĞµĞ²Ğ°Ğ¹Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ (device_map="auto" ÑĞ°Ğ¼ Ñ€Ğ°ÑĞ¿Ñ€ĞµĞ´ĞµĞ»Ğ¸Ñ‚ Ğ²ĞµÑĞ°;
            # Ğ½Ğ¾ Ğ²Ñ…Ğ¾Ğ´Ñ‹ Ğ½ÑƒĞ¶Ğ½Ğ¾ ĞºĞ¸Ğ½ÑƒÑ‚ÑŒ Ğ½Ğ° Ñ‚Ğ¾Ñ‚ Ğ¶Ğµ Ğ¿ĞµÑ€Ğ²Ñ‹Ğ¹ Ğ´ĞµĞ²Ğ°Ğ¹Ñ lm_head)
            enc = {k: v.to(device) for k, v in enc.items()}
            with torch.inference_mode():
                out = model.generate(
                    **enc,
                    forced_bos_token_id=forced_bos_id,
                    do_sample=False,
                    max_new_tokens=96
                )

            decoded = tokenizer.batch_decode(out, skip_special_tokens=True)
            decoded_all.extend(decoded)


            if num_batch%save_par_batch:
              out_df = pd.DataFrame({
                  f"{SRC_LANG}": texts[:len(decoded_all)],
                  f"{tgt_lang}": decoded_all,
              })
              out_path = f"{prefix_for_saver}{tgt_lang}.parquet"
              out_df.to_parquet(out_path, index=False)
        out_df.to_parquet(out_path, index=False)

```
